```python
P=df["Physics Marks"]   #Reading Physics Marks as Array
C=df["Chem Marks"]
B=df["Biology Marks"]
Overall_Avg=df["AVG"]
Sum=df["SUM"]
Avg=df["AVG"]        #Reading average of all 3 subjects as an Array
GPA=df["HighSchool GPA"] #Reading GPA of a Studenty as an Array
print(P)
print(type(P))       #Type of P is a pandas.series
print(C)
print(B)
print(Sum)
print(Avg)
print(GPA)

Phy_Avg=P.mean()
Chem_Avg=C.mean()
Bio_Avg=B.mean()
Overall_Avg=Avg.mean()

print("Phy_Avg",Phy_Avg)
print("Chem_Avg",Chem_Avg)
print("Bio_Avg",Bio_Avg)
print("Overall_Avg",Overall_Avg)
```

```python
import scipy
from scipy import stats  #Applying Linear Regression to find Total using Physics Score 

my_y = Sum
my_x = P

slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(my_x, my_y)
    
print(slope)
print(intercept)

plt.xlabel('Physics Marks')
plt.ylabel('Total Marks')
plt.scatter(P,Sum)
plt.show()

def calc_total_with_phy_score():
    print("Enter Score")
    n=int(input())
    ans=(n*slope)+intercept
    print(ans)
calc_total_with_phy_score()

# INFERENCE
# There is a linear relationship between Physics and Total Marks
# Students who score well in genreal perform well in Physics and vice versa
```

```python
import datetime 
Application_no=df["Application No"]
print(Application_no)
DOB=df["DOB"]
print(DOB)
print(type(DOB[0]))
def get_month_no(mon):
    if(mon=="Jan"):
      return 1
    if(mon=="Feb"):
      return 2
    if(mon=="Mar"):
      return 3
    if(mon=="Apr"):
      return 4
    if(mon=="May"):
      return 5
    if(mon=="Jun"):
      return 6
    if(mon=="Jul"):
      return 7
    if(mon=="Aug"):
      return 8
    if(mon=="Sep"):
      return 9
    if(mon=="Oct"):
      return 10
    if(mon=="Nov"):
      return 11
    if(mon=="Dec"):
      return 12
k=DOB[10]
print(k)
print(k[-4:])
print(len(k))
print(k[-4:0])
```

```python
Bday=[]                   #Converting String to Datetime
for x in DOB:
    dd=mm=yyyy=flag=0
    if(x[1]=="-"):
      dd=int(x[0])
      flag=0
    else:
      dd=int(x[0:2])
      flag=1
    yyyy=x[-4:]
    yyyy=int(yyyy)
    if(flag==0):
      mm=get_month_no(x[2:-5])
    else:
      mm=get_month_no(x[3:-5])
    elem=datetime.datetime(yyyy,mm,dd)
    Bday.append(elem)
print(Bday)
print(type(Bday[12]))
print(Bday[0].year)
# From Bday get list of all the years
```

```python
# Genreal PIE Chart syntax 
plt.style.use("bmh")
Year=[]
for x in Bday:
    Year.append(x.year)
print(Year)
print(type(Year[0]))
print(Year[0])
# Get unique year values and store it in a list called labels and sort it in ascending order
labels=[]
for x in Year:
    if(x not in labels):
      labels.append(x)
labels.sort()
print(labels)
# Get the count of each unique year and store it in a list called sizes
sizes=[]
for x in labels:
    count=0
    for y in Year:
        if(x==y):
            count+=1
    sizes.append(count)
print(sizes)
# Plot the pie chart using labels and sizes list and display the percentage of each year
plt.figure(figsize=(20,20))
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.show()
# INFERENCE
# Most of the students are born in 1998
# More than 90 percent of students are born between 1996 and 1999
```

```python
year=[]                       #Distribution Based on Birth Year
for x in Bday:
  year.append(x.year)
d={}
for x in year:
  if x not in d.keys():
    d[x]=1
  else:
    d[x]=d[x]+1
print(d)
years=list(d.keys())
count=list(d.values())
plt.bar(years,count)
plt.show()
#INFERENCE
#Time period between 1997 and 1999 has seen the most no of Students
```
